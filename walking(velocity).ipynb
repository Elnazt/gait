```python
#for n in range(len(files)):

pre_data=pd.read_csv(file)
data=pre_data.iloc[3:,:]
data=data.reset_index(drop=True)
data=data.rename(columns=data.iloc[0]).drop(data.index[0])
df=data.rename(columns={"Time,s":"time","DTS 10V ANALOG INPUT,mV":"auditory","LT DTS FOOT SWITCH,On/Off":"LFoot","RT DTS FOOT SWITCH,On/Off":"RFoot"})

#we should remove first 5 seconds and the index was reset:

#df=df.iloc[time:,:]
#df=df.reset_index(drop=True) 
df.head()
```

```python
#make the whole column of RFoot float
df=df.astype(float)
```

# eliminate the first 5 buzz and the last one from the dataset

```python
lenaud=len(df.auditory)
lenaud
```

```python
aud=pd.DataFrame()
dfaud=pd.DataFrame(df.auditory)
for n in range(lenaud-1):
    aud[n+1]=((dfaud.iloc[n+1])-(dfaud.iloc[n]))/2          
           
aud
```

```python
taud_init=aud.T
taud_init
```

```python
time1=df.time
taud=taud_init.join(time1)
taud
```

```python
arr=taud.auditory.to_numpy()

peaks, _ = find_peaks(arr, distance=800,height=1100)
peaks
```

```python
peaks[len(peaks)-2]
```

```python
start=taud.time.iloc[peaks[5]]
end=taud.time.iloc[peaks[(len(peaks)-2)]]
end
```

```python
#for x in range(lenaud-1):
#    if taud.auditory.iloc[x]>1100:
#        print(taud.iloc[x])
```

```python
#dataset from 6th buzz to last-1 buzz
trial=df.iloc[peaks[5]:peaks[-2]]
trial
```

```python
### reset index 
trial.reset_index(drop=True,inplace=True)
trial
```

# RFoot

```python
##fine the deriviation in each frame

RF=pd.DataFrame(trial.RFoot)
diff=pd.DataFrame()
for i in range(len(trial.RFoot)-1):
    diff[i+1]=((RF.iloc[i+1])-(RF.iloc[i]))/2

diff
```

```python

diff=diff.T
time=trial["time"]
diff=diff.join(time)

diff
```

```python
###making a column for onsets
trial["onset_RF"]=np.nan
```

```python
### find heel onsets for right foot
for n in range(len(trial.RFoot)-1):
    if diff.RFoot.iloc[n]>0.13 and diff.RFoot.iloc[n]<0.14:
        trial.onset_RF.iloc[n]='heel onset'

        
trial       
```

```python
## find toe onsets for the right foot :
for n in range(len(trial.RFoot)-1):
    if diff.RFoot.iloc[n]>0.53 and diff.RFoot.iloc[n]<0.54:
        trial.onset_RF.iloc[n]='toe onset'
trial
```

# LF

```python

LF=pd.DataFrame(trial.LFoot)
#diff=pd.DataFrame(columns=['velocityRF'])
Ldiff=pd.DataFrame()
for i in range(len(trial.LFoot)-1):
    Ldiff[i+1]=((LF.iloc[i+1])-(LF.iloc[i]))/2

Ldiff
```

```python
Ldiff=Ldiff.T
Ltime=trial["time"]
Ldiff=Ldiff.join(Ltime)

Ldiff
```

```python
###making a column for onsets
trial["onset_LF"]=np.nan
```

```python
###finding heel onset for left foot
for n in range(len(trial.LFoot)-1):
    if Ldiff.LFoot.iloc[n]>0.13 and Ldiff.LFoot.iloc[n]<0.14:
        trial.onset_LF.iloc[n]='heel onset'

        
trial  
```

```python
## find toe onsets for the left foot :
for f in range(len(trial.LFoot)-1):
    if Ldiff.LFoot.iloc[f]>0.53 and Ldiff.LFoot.iloc[f]<0.54:
        trial.onset_LF.iloc[f]='toe onset'
trial
```
