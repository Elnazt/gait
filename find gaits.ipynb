```python
#finding the heel onset and eliminate the first 5 and last one
Rgait_init=[]
for m in range(len(data.RFoot)):
    if data.onset_RF.iloc[m]=="heel onset":
        Rgait_init.append(data.time.iloc[m])
            
Rgait=pd.DataFrame(Rgait_init,columns=['R_Honset_time']) 
Rgait
```

```python
#finding the heel onset and eliminate the first 5 and last one
Lgait_init=[]
for m in range(len(data.RFoot)):
    if data.onset_LF.iloc[m]=="heel onset":
        Lgait_init.append(data.time.iloc[m])
            
Lgait=pd.DataFrame(Lgait_init,columns=['L_Honset_time']) 
Lgait
```

```python
###find each right gait cycle
Rgaitcyc=[]
for n in range(len(Rgait.R_Honset_time)-1):
    sub=Rgait.R_Honset_time.iloc[n+1]-Rgait.R_Honset_time.iloc[n]
    Rgaitcyc.append(sub)

Rgaitcyc=pd.DataFrame(Rgaitcyc,columns=['Rgait_cyc'])
#gaitcyc.to_csv('gaitcycle_6A.csv')
Rgaitcyc
```

```python
###find the average of right gait cycle
lenR=len(Rgaitcyc.Rgait_cyc)
Rsumm=Rgaitcyc.sum()
Rgait_ave=Rsumm/lenR

Rsumm
lenR
Rgait_ave
```

```python
#find each left gait cycle
Lgaitcyc=[]
for n in range(len(Lgait.L_Honset_time)-1):
    sub=Lgait.L_Honset_time.iloc[n+1]-Lgait.L_Honset_time.iloc[n]
    Lgaitcyc.append(sub)

Lgaitcyc=pd.DataFrame(Lgaitcyc,columns=['Lgait_cyc'])
Lgaitcyc
```

```python
#find the average of left gait cycle

lenL=len(Lgaitcyc.Lgait_cyc)
Lsumm=Lgaitcyc.sum()
Lgait_ave=Lsumm/lenL

Lsumm
lenL
Lgait_ave
```
